```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

book_title_list = []
locations_list = []
prices_list = []

for i in range (1,50):

  url = "https://www.ebay.co.uk/sch/267/i.html?_nkw=watches&_from=R40&_ipg=240&_pgn"+str(i)
  response = requests.get (url)
  soup = BeautifulSoup(response.text,'lxml')

  book_titles = soup.find_all ('div',class_='s-item__title')
  for book in book_titles:
    book_title_list.append(book.text.strip())

  locations = soup.find_all ('span',class_="s-item__location s-item__itemLocation")
  for location in locations:
    locations_list.append(location.text.strip())

  prices = soup.find_all ('span',class_='s-item__price')
  for price in prices:
    prices_list.append(price.text.strip())
```

```python
print(book_title_list)
print(locations_list)
print(prices_list)
```

```python
print(len(book_title_list))
print(len(locations_list))
print(len(prices_list))
```

```python
min_len = min(len(book_title_list),len(locations_list),len(prices_list))

book_title_list = book_title_list [:min_len]
locations_list = locations_list [:min_len]
prices_list = prices_list [:min_len]
```

```python
df=pd.DataFrame({'Books titles':book_title_list,'Country of Origin':locations_list,'Value':prices_list})
```
