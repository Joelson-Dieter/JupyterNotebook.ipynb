```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

titles_list = []
prices_list = []

for page in range (1, 100):

  url = "https://www.ebay.co.uk/b/bn_1676496?_dmd=2&mag=1&rt=nc&{page}"
  response = requests.get (url)
  soup = BeautifulSoup (response.text,'lxml')

  titles = soup.find_all ('h3',class_='textual-display bsig__title__text')
  for title in titles:
    titles_list.append(title.text.strip())

  prices = soup.find_all ('span',class_='textual-display bsig__price bsig__price--displayprice')
  for price in prices:
    prices_list.append(price.text.strip())

```

```python
print(titles_list)
print(prices_list)
```

```python
print(len(titles_list))
print(len(prices_list))
```

```python
min_len = min(len(titles_list),len(prices_list))
titles_list = titles_list [:min_len]
prices_list = prices_list [:min_len]
```

```python
df = pd.DataFrame ({'Titles':titles_list,'Prices':prices_list})
```
