```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

titles_list = []
items_location_list = []
prices_list = []

for i in range (1,100):
  url = "https://www.ebay.co.uk/sch/i.html?_nkw=watches&_sacat=0&_from=R40&rt=nc&_ipg=240&_pgn" + str (i)
  response = requests.get (url)
  soup = BeautifulSoup (response.text,'lxml')

  titles = soup.find_all ('div',class_='s-item__title')
  for title in titles:
    titles_list.append(title.text.strip())

  items_location = soup.find_all ('span',class_='s-item__location s-item__itemLocation')
  for item in items_location:
    items_location_list.append (item.text.strip())

  prices = soup.find_all ('span',class_='s-item__price')
  for price in prices:
    prices_list.append (price.text.strip())
```

```python
print(titles_list)
print(items_location_list)
print(prices_list)
```

```python
print(len(titles_list))
print(len(items_location_list))
print(len(prices_list))
```

```python
min_len = min(len(titles_list),len(items_location_list),len(prices_list))

titles_list = titles_list [:min_len]
items_location_list = items_location_list [:min_len]
prices_list = prices_list [:min_len]
```

```python
df = pd.DataFrame({'Titles':titles_list,'Country of origin':items_location_list,'Value':prices_list})
print(df)
```
