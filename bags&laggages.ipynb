```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

names_list = []
stars_list = []
reviews_list = []
discounts_list = []
prices_list = []
comments_list = []

for i in range (1, 100):
  url = "https://www.flipkart.com/all/~cs-xt03i9k5xc/pr?sid=all&collection-tab-name=Bags+and+Luggage&fm=neo%2Fmerchandising&iid=M_f60d0754-69b7-4728-8664-2b92ee45562f_1_372UD5BXDFYS_MC.X8KIDQHRP5Q0&otracker=hp_rich_navigation_8_1.navigationCard.RICH_NAVIGATION_Fashion%7EBags%2C%2BSuitcases%2B%26%2BLuggage_X8KIDQHRP5Q0&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_8_L1_view-all&cid=X8KIDQHRP5Q0&page"+str(i)

  response = requests.get (url)
  soup = BeautifulSoup (response.text,'lxml')

  names = soup.find_all ('a',class_='wjcEIp')
  for name in names:
    names_list.append(name.text.strip())

  stars = soup.find_all ('div',class_='XQDdHH')
  for star in stars:
    stars_list.append(star.text.strip())

  reviews = soup.find_all ('span',class_='Wphh3N')
  for review in reviews:
    reviews_list.append (review.text.strip())

  discounts = soup.find_all ('div',class_='UkUFwK')
  for discount in discounts:
    discounts_list.append(discount.text.strip())

  prices = soup.find_all ('div',class_='Nx9bqj')
  for price in prices:
    prices_list.append(price.text.strip())

  comments = soup.find_all ('div',class_='yiggsN O5Fpg8')
  for comment in comments:
    comments_list.append(comment.text.strip())
```

```python
print(names_list)
print(stars_list)
print(reviews_list)
print(discounts_list)
print(prices_list)
print(comments_list)
```

```python
print(len(names_list))
print(len(stars_list))
print(len(reviews_list))
print(len(discounts_list))
print(len(prices_list))
print(len(comments_list))
```

```python
min_len = min(len(names_list),len(stars_list),len(reviews_list),len(discounts_list),len(prices_list),len(comments_list))

names_list = names_list [:min_len]
stars_list = stars_list [:min_len]
reviews_list = reviews_list [:min_len]
discounts_list = discounts_list [:min_len]
prices_list = prices_list [:min_len]
comments_list = comments_list [:min_len]
```

```python
df=pd.DataFrame({'Nmaes':names_list,
                 'Stars':stars_list,
                 'Reviews':reviews_list,
                 'Discounts':discounts_list,
                 'Valu':prices_list,
                 'Comments':comments_list})
```
