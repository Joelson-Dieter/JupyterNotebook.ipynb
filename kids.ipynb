```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
from time import sleep
service = Service (r"C:\Users\Administrator\OneDrive\Documents\Data_Entry\PythonDOCS\chromedriver.exe")
driver = webdriver.Chrome (service=service)
driver.get(r"https://www.flipkart.com/")
driver.maximize_window()

input_search = driver.find_elements(By.XPATH,"(//input[@class='Pke_EE'])")

search_button = driver.find_elements(By.XPATH,"(//button[@class='_2iLD__'])")

input_search[0].send_keys("Kids,girls dresses")

sleep(5)

search_button[0].click()
```

```python
product_name = driver.find_elements(By.XPATH,"(//div[@class='syl9yP'])")
old_price = driver.find_elements(By.XPATH,"(//div[@class='yRaY8j'])")
new_price = driver.find_elements(By.XPATH,"(//div[@class='Nx9bqj'])")
discounts = driver.find_elements(By.XPATH,"(//div[@class='UkUFwK'])")
```

```python
# Find the minimum length of the lists to avoid index errors
min_length = min(len(product_name), len(old_price), len(new_price), len(discounts))

# Store data in a list first (efficient way to build DataFrame)
data = []
for i in range(min_length):  
    data.append({
        'Product name': product_name[i].text,
        'Old price': old_price[i].text,
        'New price': new_price[i].text,
        'Discounts': discounts[i].text
        
    })  

# Convert the list to a Pandas DataFrame
df_kids = pd.DataFrame(data)

# Display the DataFrame
print(df_kids)
```
