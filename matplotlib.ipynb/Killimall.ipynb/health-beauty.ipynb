```python
import requests

from bs4 import BeautifulSoup

import pandas as pd

products_list = []

reviews_list = []

prices_list= []

for i in range(1,278):
  url = "https://www.kilimall.co.ke/category/health-beauty?id=482&form=category&page"+str(i)

  response = requests.get(url)

  soup = BeautifulSoup(response.text, 'lxml')

  products = soup.find_all('p',class_='product-title')

  for product in products:

    products_list.append(product.text.strip())

  reviews = soup.find_all('span',class_='reviews')

  for review in reviews:

    reviews_list.append(review.text.strip())

  prices = soup.find_all('div',class_="product-price")

  for price in prices:

    prices_list.append(price.text.strip())

```

```python
print(products_list)

print(reviews_list)

print(prices_list)


```

```python
print(len(prices_list))
print(len(reviews_list))
print(len(prices_list))
```

```python
df=pd.DataFrame({'Product-title':products_list,'Reviews':reviews_list,'Prices':prices_list})
print(df)
```
