```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

names_list = []

reviews_list = []

prices_list = []

for page in range (1, 49):

  url = "https://www.kilimall.co.ke/search?q=boxers&page=1&source=search|hotSearch|earphone&{page}"

  response = requests.get(url)

  soup = BeautifulSoup(response.text, 'lxml')

  names = soup.find_all('p',class_="product-title")
  for name in names:
    names_list.append(name.text.strip())

  reviews = soup.find_all('span',class_="reviews")
  for review in reviews:
    reviews_list.append(review.text.strip())

  prices=soup.find_all('div',class_='product-price')

  for price in prices:
    prices_list.append(price.text.strip())
```

```python
print(names_list)

print(reviews_list)

print(reviews_list)
```

```python
print(len(names_list))

print(len(reviews_list))

print(len(prices_list))
```

```python
df=pd.DataFrame({'Names':names_list, 'Reviews':reviews_list,'Prices':prices_list})

print(df)
```
