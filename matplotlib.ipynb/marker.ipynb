```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([20,30,40,50,60])
y=np.array([100,105,110,120,130])

plt.plot(x,y)
plt.title("Daily Routine Treatment",color='r',size=10)
plt.xlabel("Blood sugar",color='r',size='10')
plt.ylabel("Average Calorie Burnage",color='r',size=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([10,30,10,30,10,25])
y=np.array([20,30,40,50,60,70])
plt.plot(x,y)
plt.title("End term Performance",color='y',size=20,loc='left')
plt.xlabel("Average marks",color='y',size=20)
plt.ylabel("Minimum Markd",color='y',size=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([40,30,50,60,70,20,35])
y=np.array([50,55,60,65,70,75,80,])
plt.title("Sports Watch",loc='right')
plt.xlabel("Average Pulse",color='g',size=10)
plt.ylabel("Calorie Burnage",color='g',size=10)
plt.plot(x,y,ls='dotted',lw=2,c='b',marker='o',ms=5,mec='r',mfc='y')
plt.show()
```
