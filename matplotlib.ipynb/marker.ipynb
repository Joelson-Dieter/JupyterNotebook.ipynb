```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array ([2,3,5,7,])
plt.plot (ypoint,marker = '*')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoint = np.array ([3,8,1,10])
plt.plot(ypoint,marker = 'o',ms=15,mec='red',mfc='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([2,4,5,8,2])
plt.plot(ypoints,marker='o',ms=10,mec='g',mfc='m')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array ([3,8,1,10,8])
plt.plot(ypoints,marker='o',mec='r',mfc='g',ms=10.5,ls='dotted',lw='5.7',c='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array ([3,8,1,10])
plt.plot(ypoints,marker='*',ms=20,mec='r',mfc='b',ls='dashed',lw='1.5',c='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y1 = np.array ([3,8,1,10])
y2 = np.array ([6,2,7,11])

plt.plot(y1,c='y',ls='dashed')
plt.plot(y2,c='r',ls='dotted')

plt.show()
```
