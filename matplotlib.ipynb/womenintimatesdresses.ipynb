```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

titles_list = []

prices_list = []

comments_list = []
```

```python

url = "https://www.ebay.com/b/Womens-Intimates-Sleep/11514/bn_661761"
response = requests.get (url)

soup = BeautifulSoup (response.text,'lxml')

```

```python
titles = soup.find_all ('h3',class_='textual-display bsig__title__text')
for title in titles:
  titles_list.append(title.text.strip())
print(titles_list)
```

```python
prices = soup.find_all ('span',class_='textual-display bsig__price bsig__price--displayprice')
for price in prices:
  prices_list.append(price.text.strip())
print(prices_list)
```

```python
comments = soup.find_all ('span',class_='textual-display bsig__generic bsig__logisticsCost')
for comment in comments:
  comments_list.append(comment.text.strip())
print(comments_list)
```

```python
print(len(titles_list))

print(len(prices_list))

print(len(comments_list))
```

```python
min_len = min(len(titles_list), len(prices_list),len(comments_list))

titles_list = titles_list[:min_len]

prices_list = prices_list[:min_len]

comments_list = comments_list[:min_len]

df = pd.DataFrame({'Title': titles_list, 'Price': prices_list, 'Comments': comments_list})

print(df)

df
```
