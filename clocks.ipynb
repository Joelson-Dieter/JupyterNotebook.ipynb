```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

titles_list = []
descriptions_list = []
old_price_list = []
new_price_list = []
discounts_list = []
stars_list = []
reviews_list = []
```

```python
for i in range (1, 20):

  url = "https://www.flipkart.com/home-decor/clocks/pr?sid=arb%2Ckjw&otracker=categorytree&fm=neo%2Fmerchandising&iid=M_6f716224-4099-4e68-b0ba-0a6e367d1abf_1_372UD5BXDFYS_MC.AEAP3WI0UC2Q&otracker=hp_rich_navigation_5_1.navigationCard.RICH_NAVIGATION_Home%2B%26%2BFurniture%7EHome%2BDecor%7EClocks_AEAP3WI0UC2Q&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_5_L2_view-all&cid=AEAP3WI0UC2Q&page"+str(i)

  response = requests.get(url)
  soup = BeautifulSoup(response.text,'lxml')

  titles = soup.find_all('a',class_='wjcEIp')
  for title in titles:
    titles_list.append(title.text.strip())


  descriptions = soup.find_all ('div',class_='NqpwHC')
  for description in descriptions:
    descriptions_list.append(description.text.strip())

  old_price = soup.find_all ('div',class_='yRaY8j')
  for price in old_price:
    old_price_list.append(price.text.strip())

  new_price = soup.find_all ('div',class_='Nx9bqj')

  for price in new_price:
    new_price_list.append(price.text.strip())

  discounts = soup.find_all ('div',class_='UkUFwK')
  for discount in discounts:
    discounts_list.append(discount.text.strip())

  stars = soup.find_all ('div',class_='XQDdHH')
  for star in stars:
    stars_list.append(star.text.strip())

  reviews = soup.find_all ('span',class_='Wphh3N')
  for review in reviews:
    reviews_list.append(review.text.strip())
```

```python
min_len = min(len(titles_list), len(descriptions_list),len(old_price_list),len(new_price_list),len(discounts_list),len(stars_list),len(reviews_list))

titles_list = titles_list[:min_len]
descriptions_list = descriptions_list[:min_len]
old_price_list = old_price_list[:min_len]
new_price_list = new_price_list[:min_len]

discounts_list = discounts_list[:min_len]

stars_list = stars_list[:min_len]

reviews_list = reviews_list[:min_len]
```

```python
df=pd.DataFrame({'Titles':titles_list,
                 'Description':descriptions_list,
                 'Old price':old_price_list,
                 'New price':new_price_list,
                 'Discounts':discounts_list,
                 'Stars':stars_list,
                 'Reviews':reviews_list
                 })

print(df)
```
